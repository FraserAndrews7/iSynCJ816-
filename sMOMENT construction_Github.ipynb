```python
import re

def add_coefficients(r,coefficient):
    coefficient=float(coefficient)
    r=str(r)
    model6.reactions.get_by_id(r).add_metabolites({model6.metabolites.get_by_id("prot_pool"):-coefficient})
    #print(model6.reactions.get_by_id(r))

def extract_coefficient(reaction_string, metabolite_name):
    # Define a regular expression pattern to match the coefficient and the metabolite name
    pattern = r'([\d.e-]+)\s*' + re.escape(metabolite_name)
    
    # Search for the pattern in the reaction string
    match = re.search(pattern, reaction_string)
    
    if match:
        # Extract and return the coefficient
        coefficient = match.group(1)
        return float(coefficient)
    else:
        # If the metabolite is not found, return 0
        return 0.0

# Example usage
reaction_string = "nadph_c + 3.286491602096798e-05 prot_pool + pyr_c --> lac__L_c + nadp_c"
metabolite_name = "prot_pool"
coefficient = extract_coefficient(reaction_string, metabolite_name)
print(f"The coefficient of {metabolite_name} is: {coefficient}")
```
